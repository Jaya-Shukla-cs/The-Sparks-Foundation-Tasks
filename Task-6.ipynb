## Step 1 : Importing the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2 : Loading the dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
X.head()
```

```python
X.tail()
```

```python
X.info()
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
y = iris.target
y
```

## Step 3 : Splitting the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

## Step 4 : Creating object of the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtf = DecisionTreeClassifier()
dtf.fit(X_train, y_train)
```

```python
y_predict = dtf.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

## Step 5 : Visualizing the decision tree

```python
from sklearn import tree
```

```python
fn = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
cn = ['setosa', 'versicolor', 'virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtf, feature_names = fn, class_names = cn, filled = True);
```
